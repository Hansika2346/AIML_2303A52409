```python
import matplotlib.pyplot as plt
rider = ["sai","anu","manu","anu","akki"]
age = [40,32,29,25,18]
weight = [68,62,59,50,43]
height = [150,131,120,169,178]
plt.scatter(weight,height)
plt.title("Relationship between weiht and height of riders")
plt.ylabel("Height in Meters")
plt.xlabel("weight of riders in kgs")
plt.show()
```

```python
plt.figure(figsize=(18,4))
plt.title("riders's weight in kgs",fontsize=18, fontstyle='italic')

plt.bar(x=rider,height=weight,width=0.5,color="orange",edgecolor="k",alpha=0.7)
plt.xlabel("rider",fontsize=12)
plt.yticks(fontsize=13,rotation=30)
plt.ylabel("weight (in kgs)",fontsize=18)
plt.show()

```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='yellow',edgecolor='k', alpha=0.75,bins=5)

plt.title("Histogram of rider weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
